```python
from tkinter import *
import math as m
window = Tk()
window.title("Simple Calculator")

Entryfeild = Entry(window,width = 20,borderwidth = 5,font = ('Times',20) )
Entryfeild.grid(row=0,column=0,columnspan = 4,padx = 15,pady = 15)

def button_click(number):
    current = Entryfeild.get()
    Entryfeild.delete(0,END)
    Entryfeild.insert(0,str(current)+str(number))
    
def button_clear():
    Entryfeild.delete(0,END)
    
def button_equals():
    Second_number = Entryfeild.get()    
    Entryfeild.delete(0,END)
    if math == "addition":
        Entryfeild.insert(0,f_num+ int(Second_number))
    if math == "subtraction":
        Entryfeild.insert(0,f_num- int(Second_number))
    if math == "multiplication":
        Entryfeild.insert(0,f_num* int(Second_number))
    if math == "divison":
        Entryfeild.insert(0,f_num / int(Second_number))
    
def button_add():
    first_number = Entryfeild.get()
    global f_num
    global math
    math = "addition"
    f_num = int(first_number)
    Entryfeild.delete(0,END)
    
def button_subtract():
    first_number = Entryfeild.get()
    global f_num
    global math
    math = "subtraction"
    f_num = int(first_number)
    Entryfeild.delete(0,END)

def button_multiply():
    first_number = Entryfeild.get()
    global f_num
    global math
    math = "multiplication"
    f_num = int(first_number)
    Entryfeild.delete(0,END)

def button_divison():
    first_number = Entryfeild.get()
    global f_num
    global math
    math = "divison"
    f_num = int(first_number)
    Entryfeild.delete(0,END)
    
    

button_1 = Button(window,text = '1',width = 5,font = ('Times',20),command =lambda: button_click(1))
button_2 = Button(window,text = '2',width = 5,font = ('Times',20),command =lambda: button_click(2))
button_3 = Button(window,text = '3',width = 5,font = ('Times',20),command =lambda: button_click(3))
button_4 = Button(window,text = '4',width = 5,font = ('Times',20),command =lambda: button_click(4))
button_5 = Button(window,text = '5',width = 5,font = ('Times',20),command =lambda: button_click(5))
button_6 = Button(window,text = '6',width = 5,font = ('Times',20),command =lambda: button_click(6))
button_7 = Button(window,text = '7',width = 5,font = ('Times',20),command =lambda: button_click(7))
button_8 = Button(window,text = '8',width = 5,font = ('Times',20),command =lambda: button_click(8))
button_9 = Button(window,text = '9',width = 5,font = ('Times',20),command =lambda: button_click(9))
button_0 = Button(window,text = '0',width = 5,font = ('Times',20),command =lambda: button_click(0))

button_div = Button(window,text = '/',width = 5,font = ('Times',20),command = button_divison)
button_clear = Button(window,text = 'C',width = 5,font = ('Times',20),command = button_clear)
button_equal = Button(window,text = '=',width = 5,font = ('Times',20),command = button_equals)
button_plus = Button(window,text = '+',width = 5,font = ('Times',20),command = button_add)
button_minus = Button(window,text = '-',width = 5,font = ('Times',20),command = button_subtract)
button_multiply = Button(window,text = 'x',width = 5,font = ('Times',20),command = button_multiply)



button_1.grid(row=3,column = 0,padx=2,pady=5)
button_2.grid(row=3,column = 1,padx=2,pady=5)
button_3.grid(row=3,column = 2,padx=2,pady=5)
button_plus.grid(row=3,column=3,padx=2,pady=5)

button_4.grid(row=2,column = 0,padx=2,pady=5)
button_5.grid(row=2,column = 1,padx=2,pady=5)
button_6.grid(row=2,column = 2,padx=2,pady=5)
button_minus.grid(row = 2,column = 3,padx=2,pady=5)

button_7.grid(row=1,column = 0,padx=2,pady=5)
button_8.grid(row=1,column = 1,padx=2,pady=5)
button_9.grid(row=1,column = 2,padx=2,pady=5)
button_multiply.grid(row=1,column=3,padx=2,pady=5,)

button_clear.grid(row=4,column=0,padx=2,pady=5)
button_0.grid(row=4,column=1,padx=2,pady=5)
button_div.grid(row=4,column=2,padx=2,pady=5)
button_equal.grid(row = 4,column=3,padx=2,pady=5)



window.mainloop()
```
